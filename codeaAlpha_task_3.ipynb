```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

```

```python
df = pd.read_csv("/content/archive (3).zip")
```

```python
print(df.head())
print(df.info())
print(df.shape)
```

```python
os.makedirs("charts", exist_ok=True)
```

```python
sns.set_style("whitegrid")

# 1. Content Type Distribution
plt.figure(figsize=(6,4))
df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("charts/content_type.png")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Titles")
plt.tight_layout()
plt.savefig("charts/top_countries.png")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df["release_year"].value_counts().sort_index().tail(20).plot()
plt.title("Content Released Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("charts/release_year_trend.png")
plt.show()
```

```python
plt.figure(figsize=(7,4))
df["rating"].value_counts().head(10).plot(kind="bar")
plt.title("Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("charts/content_rating.png")
plt.show()
```

```python
top5 = df["type"].value_counts()

plt.figure(figsize=(5,5))
plt.pie(
    top5,
    labels=top5.index,
    autopct="%1.1f%%"
)
plt.title("Movies vs TV Shows Share")
plt.tight_layout()
plt.savefig("charts/type_pie_chart.png")
plt.show()
```
